```python
#Monday
#Mean Squared Error, Mean Absolute Error, Mean Absolute Percentage Error
x_monday = monday['hour'].values.reshape(-1, 1)
y_monday = monday['monday'].values
poly = PolynomialFeatures(degree=15)
x_15_monday = poly.fit_transform(x_monday)
x_15_monday.shape
linear15_monday = linear_model.LinearRegression()
linear15_monday.fit(x_15_monday, y_monday)
print("Monday")
(metrics.mean_squared_error(y_monday[-4:], linear15_monday.predict(x_15_monday[-4:])),
metrics.mean_absolute_error(y_monday[-4:], linear15_monday.predict(x_15_monday[-4:])),
metrics.mean_absolute_percentage_error(y_monday[-4:], linear15_monday.predict(x_15_monday[-4:])))
```

## 4. With `saturday`, use `train_test_split` to create training and test sets and build a model. Create predictions using the xtest from and determine the errors between these predictions and the ytest (MSE, MAE, MAPE).

## repeat for `monday`

```python
from sklearn.model_selection import train_test_split

x = saturday['hour'].values.reshape(-1, 1)
y = saturday['saturday'].values

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
#linear = linear_model.LinearRegression().fit(xtrain, ytrain)

xtrain5 = PolynomialFeatures(degree=5).fit_transform(xtrain)
xtest5 = PolynomialFeatures(degree=5).fit_transform(xtest)
linear5 = linear_model.LinearRegression().fit(xtrain5, ytrain)

xtrain15 = PolynomialFeatures(degree=15).fit_transform(xtrain)
xtest15 = PolynomialFeatures(degree=15).fit_transform(xtest)
linear15 = linear_model.LinearRegression().fit(xtrain15, ytrain)

xtrain20 = PolynomialFeatures(degree=20).fit_transform(xtrain)
xtest20 = PolynomialFeatures(degree=20).fit_transform(xtest)
linear20 = linear_model.LinearRegression().fit(xtrain20, ytrain)

size = 8
plt.scatter(xtest, ytest, c='black', s=size)
plt.scatter(xtest, linear5.predict(xtest5), c='r', s=size)
plt.scatter(xtest, linear15.predict(xtest15), c='b', s=size)
plt.scatter(xtest, linear20.predict(xtest20), c='g', s=size)

print("MSE 5, 15, 20")
print(metrics.mean_squared_error(ytest[-4:], linear5.predict(xtest5[-4:])),
      metrics.mean_squared_error(ytest[-4:], linear15.predict(xtest15[-4:])),
      metrics.mean_squared_error(ytest[-4:], linear20.predict(xtest20[-4:])))
print("MAE 5, 15, 20")
print(metrics.mean_absolute_error(ytest[-4:], linear5.predict(xtest5[-4:])),
metrics.mean_absolute_error(ytest[-4:], linear15.predict(xtest15[-4:])),
metrics.mean_absolute_error(ytest[-4:], linear20.predict(xtest20[-4:])))
print("MAPE 5, 15, 20")
print(metrics.mean_absolute_percentage_error(ytest[-4:], linear5.predict(xtest5[-4:])),
metrics.mean_absolute_percentage_error(ytest[-4:], linear15.predict(xtest15[-4:])),
metrics.mean_absolute_percentage_error(ytest[-4:], linear20.predict(xtest20[-4:])))
```
